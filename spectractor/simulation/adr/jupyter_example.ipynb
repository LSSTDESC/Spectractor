```python
import matplotlib.pyplot as plt
import adr_utils as adru
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['image.origin'] = True
```

```python
lbda_ref = 4000
lbdas = np.linspace(lbda_ref,6000,51)
fitsfile = 'reduc_20170530_130.fits'
```

```python
lat,_ = adru.get_lat_tel('CTIO')

meadr = adru.instanciation_adr(fitsfile, lat, lbda_ref)
```

```python
max_x, max_y = adru.get_max_shift_for_lbdas(meadr, lbdas)
print('Maximum shifts are {:.3f} arcsec in x and {:.3f} arcsec in y'.format(max_x,max_y))

max_x_pix, max_y_pix = adru.in_pixel(max_x, fitsfile), adru.in_pixel(max_y, fitsfile)
print('                   {:.3f} pixels          {:.3f} pixels'.format(max_x_pix,max_y_pix))
```

```python
xs, ys = adru.get_adr_shift_for_lbdas(meadr, lbdas)
xs_pix, ys_pix = adru.in_pixel(xs, fitsfile), adru.in_pixel(ys, fitsfile)

adru.plot_adr(lbdas, xs_pix, ys_pix, units='pixels')
```
